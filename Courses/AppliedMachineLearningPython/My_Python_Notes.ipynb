#### While

```python
"""
La estructura de While es la siguiente:

while condition:
    something
"""
# https://es.wikipedia.org/wiki/Serie_de_Leibniz
i = 0
pi_approximation = 0
pi_approximations = []
while np.abs(4*pi_approximation-np.pi)/np.pi >= 0.000001: #Umbral de precisión
    pi_approximation += ((-1)**i) / ((2*i)+1)
    pi_approximations.append(4*pi_approximation)
    #residuo = np.pi-4*pi_approximation
    i+=1
print(f'Pi Approximation: {round(4*pi_approximation, 5)}')
print(f'Iterations: {len(pi_approximations)}')
```

### Mapeo

> 1. map(); Recibe una funcion y se la aplica a los elem de la columna
2. apply(); Se ocupa si queremos transformar el DataFrame completo, recibre una función y se la aplica a c/fila del Datafram


```python
# https://archive.ics.uci.edu/ml/datasets/Adult
cols = ['age','workclass','fnlwgt','education','education_level','marital','occupation','relationship','race','sex','capital_gain','capital_loss','hours','native','target']
df = ps.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
df.columns = cols
df.head(4) # vemos los primero 4 elementos
```

```python
def remean_age(row):
  row.age = row.age - df.age.mean()
  return row

df.apply(remean_age, axis='columns')
# Si en aplly ocupamos axis='index' entonces el applyrealiza las transformaciones por fila
# Para hacer el cambio permanente debemos de poner: df =  df.apply(...)
```

### Gráficos de barras y Mapas de calor

```python
# :11999, las primeras 11999 filas
# ,: seleccionamos todas las columnas
df_month = df.loc[:11999,:].copy(deep=True) # Tomaremos las primeras 12,000 filas del DataFrame para reducir tiempos de ejecución.
# El deep true indica que quiere elementos diferentes
```

Supongamos que cada mil filas corresponden a las operaciones de un mes...

```python

# Definimos un mapeo que transforma el índice en el mes correspondiente
def month_map(index_value):
    month_value = (index_value // 1000) + 1
    return month_value

df_month['Month'] = df_month.index.map(lambda t: month_map(t))
df_month.Month.value_counts()
```

```python
# Agrupamos los valores de la colmuna V1 por mes y calculamos mínimo, máximo y promedio en cada mes.
df_view = df_month.groupby(['Month']).V1.agg([min,max,'mean'])
df_view
```

```python
# "aplanamos" el DataFrame, es decir, pasamos de un multindex a un índice
df_view = df_view.reset_index() 
df_view
```

```python
# Ejemplo de gráfico de barras --- sns.barplot()
plt.figure(figsize=(10,7))
plt.title("Promedio de V1 por Mes")
sns.barplot(x=df_view.Month, y=df_view['mean'])
plt.xlabel("Mes")
plt.ylabel("Valor")
```

```python
# Agrupamos los valores de la lista de columnas por mes y aplicamos mean
df_view = df_month.groupby(['Month'])[['V1','V2','V3','V4','V5','V6', 'V7', 'V8', 'V9','V10']].agg(['mean'])
df_view
```

```python
# Renombramos las columnas 
df_view.columns = df_view.columns.to_flat_index().str.join('_')
df_view
```

Un mapa de valores nos ayuda a comparar los valores de varios atributos, se recomienda que tu eje y sea tu variable categorica y el eje X los valores numericos

```python
# Ejemplo de mapa de calor
plt.figure(figsize=(14,7))
plt.title("Promedios de V1 a V10 por mes")
sns.heatmap(data=df_view, annot=True)
plt.xlabel("Atributos")
plt.ylabel("Mes")
```
